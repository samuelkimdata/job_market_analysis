## companies.csv & employee_counts.csv merge

```python
# Import CSV's
employee_counts = pd.read_csv("/Users/sangkim/Desktop/demo_day_project/job_market_analysis/Starter Files/employee_counts.csv")
```

```python
# Confirm CSV's are properly imported
employee_counts.head()
```

```python
# Merge employee_counts & companies on 'company_id'
company_info = pd.merge(companies, employee_counts, on = 'company_id', how = 'inner')
```

```python
# Remove all duplicates from company_id
company_info.drop_duplicates('company_id')
```

```python
# Export cleaned csv
company_info.to_csv('data_cleanup.csv', index = False)
```
