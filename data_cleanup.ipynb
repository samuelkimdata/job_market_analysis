## job_postings.csv cleanup

```python
# Import Dependencies
import pandas as pd
from datetime import datetime
```

```python
# Read in CSV
job_postings = pd.read_csv("/Users/sangkim/Desktop/demo_day_project/job_postings.csv")
```

```python
# Display first 5 rows of CSV
job_postings.head()
```
