## job_postings.csv cleanup

```python
# Import Dependencies
import pandas as pd
from datetime import datetime
```

```python
# Read in CSV
job_postings = pd.read_csv("/Users/sangkim/Desktop/demo_day_project/job_postings.csv")
```

```python
# Display first 5 rows of CSV
job_postings.head()
```

```python
# Drop 'description' and 'scraped' for irrelevency
columns_to_drop = ['description', 'scraped']
job_postings.drop(columns = columns_to_drop, inplace = True)
```

```python
# Confirm correct columns were dropped
job_postings.head()
```

```python
# Convert all time columns to timestamp
job_postings['original_listed_time'] = pd.to_datetime(job_postings['original_listed_time'], unit='ms', errors='coerce')
job_postings['closed_time'] = pd.to_datetime(job_postings['closed_time'], unit='ms', errors='coerce')
job_postings['listed_time'] = pd.to_datetime(job_postings['listed_time'], unit='ms', errors='coerce')
job_postings['expiry'] = pd.to_datetime(job_postings['expiry'], unit='ms', errors='coerce')
job_postings
```

## companies.csv & employee_counts.csv merge

```python
# Import CSV's
employee_counts = pd.read_csv("/Users/sangkim/Desktop/demo_day_project/job_market_analysis/Starter Files/employee_counts.csv")
```

```python
# Confirm CSV's are properly imported
employee_counts.head()
```

```python
# Merge employee_counts & companies on 'company_id'
company_info = pd.merge(companies, employee_counts, on = 'company_id', how = 'left')
company_info.head()
```

```python
# Remove all duplicates from company_id
cleaned_company_info = company_info.drop_duplicates('company_id')
cleaned_company_info.head()
```

```python
# Change time_recorded to datetime format
cleaned_company_info['time_recorded'] = pd.to_datetime(cleaned_company_info['time_recorded'], unit = 's', errors='coerce')
cleaned_company_info.head()
```

```python
# Export cleaned csv
cleaned_company_info.to_csv('Updated Files/company_info.csv', index = False)
```
