## Convert all salaries to yearly, then fill 'med_salary' column with results.

```python
# Indicate all values in 'pay_period' column
job_postings['pay_period'].unique()
```

```python
# Create separate dataframes for each pay_period
monthly = job_postings[job_postings['pay_period'] == 'MONTHLY']
hourly = job_postings[job_postings['pay_period'] == 'HOURLY']
yearly = job_postings[job_postings['pay_period'] == 'YEARLY']
weekly = job_postings[job_postings['pay_period'] == 'WEEKLY']
once = job_postings[job_postings['pay_period'] == 'ONCE']
```

```python
# Calculate median salary for rows where it's not already filled
hourly.loc[hourly['med_salary'].isnull(), 'med_salary'] = (
    hourly['max_salary'] + hourly['min_salary']) / 2

weekly.loc[weekly['med_salary'].isnull(), 'med_salary'] = (
    weekly['max_salary'] + weekly['min_salary']) / 2

monthly.loc[monthly['med_salary'].isnull(), 'med_salary'] = (
    monthly['max_salary'] + monthly['min_salary']) / 2

yearly.loc[yearly['med_salary'].isnull(), 'med_salary'] = (
    yearly['max_salary'] + yearly['min_salary']) / 2

once.loc[once['med_salary'].isnull(), 'med_salary'] = (
    once['max_salary'] + once['min_salary']) / 2
```

```python
# Update the 'med_salary' column in the original DataFrame with the values from the individual DataFrames
job_postings.update(hourly[['med_salary']])
job_postings.update(weekly[['med_salary']])
job_postings.update(monthly[['med_salary']])
job_postings.update(yearly[['med_salary']])
job_postings.update(once[['med_salary']])
job_postings
```

```python
# Change 'remote_allowed' and 'sponsored' columns to bool
job_postings['remote_allowed'].fillna(0, inplace = True)
job_postings['remote_allowed'] = job_postings['remote_allowed'].astype(bool)
job_postings['sponsored'] = job_postings['sponsored'].astype(bool)
print(job_postings['remote_allowed'])
print(job_postings['sponsored'])
```

```python
job_postings.to_csv("Updated Files/job_postings_cleaned.csv", index = False)
```

## Merge 'benefits.csv' with 'company_id' and 'name' from 'cleaned_company_info'

```python
# Read in 'benefits.csv' and merge with 'company_id' from 'cleaned_company_info'
benefits = pd.read_csv("/Users/sangkim/Desktop/demo_day_project/job_market_analysis/Starter Files/benefits.csv")

# Change benefits.job_id column type to str
benefits['job_id'] = benefits['job_id'].astype(str)

# Select 'job_id' and 'company_id' from job_postings dataframe, then merge with benefits dataframe on 'job_id'
job_postings_id = job_postings[['job_id', 'company_id']]
benefits_merged = pd.merge(benefits, job_postings_id, on = 'job_id', how = 'left')

# Merge benefits_merged with 'name' from 'cleaned_company_info'
company_name = cleaned_company_info[['company_id', 'name']]
benefits_name = pd.merge(benefits_merged, company_name, on = 'company_id', how = 'left')

# Reorder new dataframe
new_columns = ['job_id',
               'company_id',
               'name',
               'type']
benefits_cleaned = benefits_name[new_columns]
benefits_cleaned

# Export 'benefits_cleaned' to csv
benefits_cleaned.to_csv("Updated Files/benefits.csv", index = False)
```
